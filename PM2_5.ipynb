```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#vcituvanje na data bazata
data = pd.read_csv('PM25.csv')
```

```python
#definiranje na promenlivi na podatocite
centar = data['Centar']
gazibaba = data['GaziBaba']
karpos = data['Karpos']
lisice = data['Lisice']
miladinovci = data['Miladinovci']
mobile = data['Mobile']
rektorat = data['Rektorat']
```

```python
#Grafik so golemina 10 na 6
plt.figure(figsize=(10, 6))
#Dodavanje na Linii na sekoja lokacija na grafikot
plt.plot(centar, label='Centar')
plt.plot(gazibaba, label='GaziBaba')
plt.plot(karpos, label='Karpos')
plt.plot(lisice, label='Lisice')
plt.plot(miladinovci, label='Miladinovci')
plt.plot(mobile, label='Mobile')
plt.plot(rektorat, label='Rektorat')
#  Podnesuvanje na oznaki i naslov na grafikot
plt.xlabel('id')
plt.ylabel('PM2,5 Pollution')
plt.title('PM2,5 Pollution Comparison')
#Dodavanje Upatstvo za Sekoja od Naselbite so koja boja e oznacena na levata strana
plt.legend()
#prikazuvanje na mrezata na grafikot
plt.grid(True)
#prikazuvanje na grafikot
plt.show()
```
